```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 1000  # Sampling frequency (Hz)
N = 1024   # FFT points for DFT
t = np.arange(0, 1, 1/fs)
```

```python

def plot_time_and_freq(signal, title, fs, N):
    """Helper function: plot Time-domain, DTFT, and DFT"""
    # Time domain
    plt.figure(figsize=(14,8))
    plt.subplot(3,1,1)
    plt.plot(t[:len(signal)], signal)
    plt.title(f"Time Domain: {title}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")

    # (c) DTFT (continuous spectrum approximation using np.fft.fft with high resolution)
    w = np.linspace(-np.pi, np.pi, 4096)
    # DTFT approx = sum(signal[n]*exp(-jwn))
    # Use FFT shift trick for better visualization
    dtft = np.fft.fftshift(np.fft.fft(signal, 4096))
    freqs_dtft = np.linspace(-fs/2, fs/2, len(dtft))

    plt.subplot(3,1,2)
    plt.plot(freqs_dtft, np.abs(dtft))
    plt.title("DTFT (Approx. Continuous Spectrum)")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude")

    # (d) DFT
    dft = np.fft.fft(signal, N)
    freqs_dft = np.fft.fftfreq(N, 1/fs)
    plt.subplot(3,1,3)
    plt.stem(freqs_dft[:N//2], np.abs(dft[:N//2]), basefmt=" ")
    plt.title("DFT (Discrete Spectrum)")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude")

    plt.tight_layout()
    plt.show()
```

Question 1

```python
f = 5  # Hz
signal1 = np.sin(2*np.pi*f*t)
plot_time_and_freq(signal1, "5 Hz Sinusoid", fs, N)
```

Question 2

```python
signal2 = 1.0*np.sin(2*np.pi*5*t) + 0.5*np.sin(2*np.pi*20*t) + 0.8*np.sin(2*np.pi*50*t)
plot_time_and_freq(signal2, "Composite Signal (5Hz + 20Hz + 50Hz)", fs, N)
```

Question 3

```python
alpha = 5   # Decay rate
signal3 = np.exp(-alpha*t)  # exponential decay
plot_time_and_freq(signal3, "Exponentially Decaying Signal", fs, N)
```

Question 4

```python
pulse = np.zeros_like(t)
pulse[200:400] = 1.0  # Rectangular pulse between 0.2s and 0.4s
plot_time_and_freq(pulse, "Rectangular Pulse", fs, N)
```
